# Пропуски и очистка
Я посмотрел файл и "None" считается как NaN (это 10 строка).

```python
df.dropna().shape
```

df.dropna() просто удалил эти 91 строк

```python
df.fillna( "No education").head(11)
```

В этом случае лучше заменить константой по типу "No education". Как минимум, потому что это строковое значение (на него надо, наверное, писать свои среднее и медиану). Среднее или медиана, помоему, лучше использовать, если это надо сохранить какое-то колличество данных, которые не нулевые.

# Мода
Если считаем, что "None" -- это не полученное значение, то можно заменить модой.

```python

df.fillna({"parental_education_level": df['parental_education_level'].mode()[0]}).head(11)
```
